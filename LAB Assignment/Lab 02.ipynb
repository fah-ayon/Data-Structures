```python
#You must run this cell to print matrix and for the driver code to work
def print_matrix(m):
  row,col = m.shape
  for i in range(row):
    c = 1
    print('|', end='')
    for j in range(col):
      c += 1
      if(len(str(m[i][j])) == 1):
        print(' ',m[i][j], end = '  |')
        c += 6
      else:
        print(' ',m[i][j], end = ' |')
        c += 6
    print()
    print('-'*(c-col))
```

#Zigzag Walk

```python
def walk_zigzag(floor):
  row, col = floor.shape
  for i in range(col):
    if i%2==0:
      for j in range(0,row,2):
        print(floor[j][i], end= ' ')
    else:
      if row%2==0:
        for j in range(row-1,-1,-2):
          print(floor[j][i], end= ' ')
      else:
        for j in range(row-2,-1,-2):
          print(floor[j][i], end= ' ')
    print()

floor = np.array([[ '3' , '8' , '4' , '6' , '1'],
                  ['7' , '2' , '1' , '9' , '3'],
                  ['9' , '0' , '7' , '5' , '8'],
                  ['2' , '1' , '3' , '4' , '0'],
                  ['1' , '4' , '2' , '8' , '6']]
                )

print_matrix(floor)
print('Walking Sequence:')
walk_zigzag(floor)
#This should print
# 3 9 1
# 1 2
# 4 7 2
# 4 9
# 1 8 6
print('################')
floor = np.array([[ '3' , '8' , '4' , '6' , '1'],
                  ['7' , '2' , '1' , '9' , '3'],
                  ['9' , '0' , '7' , '5' , '8'],
                  ['2' , '1' , '3' , '4' , '0']]
                )

print_matrix(floor)
print('Walking Sequence:')
walk_zigzag(floor)
#This should print
# 3 9
# 1 2
# 4 7
# 4 9
# 1 8
```

#Wall Up Trost District

```python
def create_fence(district, depth):
    district_row, district_col = district.shape
    new_row = district_row+ (depth*2)
    new_col = district_col + (depth*2)
    new_mat = np.zeros((new_row,new_col), dtype= int)
    for i in range(new_row):
      for j in range(new_col):
          if new_mat[i][j]==0:
            new_mat[i][j]=8
    for i in range(district_row):
      for j in range(district_col):
        new_mat[i+depth][j+depth]= district[i][j]
    return new_mat



depth = 1
district = np.array([[2,3,4], [3,4,6], [2,1,4]])
print_matrix(district)
ans = create_fence(district, depth)
print_matrix(ans)
#This will print
# |  8  |  8  |  8  |  8  |  8  |
# -------------------------------
# |  8  |  2  |  3  |  4  |  8  |
# -------------------------------
# |  8  |  3  |  4  |  6  |  8  |
# -------------------------------
# |  8  |  2  |  1  |  4  |  8  |
# -------------------------------
# |  8  |  8  |  8  |  8  |  8  |
# -------------------------------
print('################')
depth = 2
district = np.array([
                 [2,3,4,1],
                 [3,4,6,5],
                 [2,1,4,7]
                ])
print_matrix(district)
ans = create_fence(district, depth)
print_matrix(ans)
```

#Crows vs Cats

```python
def strength_difference(clubroom):
  row,col = clubroom.shape
  strength_diff = np.zeros(100, dtype = int)
  total_elem = row*col
  strength_diff = np.zeros(int((total_elem-row)/2), dtype= int) #resized array
  k=0
  for i in range(row):
    for j in range(col):
      if i<j:
        strength_diff[k]= clubroom[i][j]-clubroom[j][i]
        k+=1
  return strength_diff

clubroom = np.array([
[1,  2,  9,  7],
[4,  5,  1,  8],
[3,  6,  2,  7],
[2,  8,  6,  3]
])
print_matrix(clubroom)
returned_value = strength_difference(clubroom)
print('Strength Difference Array is : ', returned_value)
unittest.output_test(returned_value, np.array([-2, 6, 5, -5, 0, 1]))
```

#ATM's Triangle

```python
def atm_triangle(n):
    new= np.zeros((n,n), dtype=int)
    row,col = new.shape
    for i in range(row):
      j = 0
      new[i][j]=new[i][i]=i+1
      if i>=2:
        for j in range(1,i):
          a=0
          while a<=j:
            new[i][j]+=new[i-1][a]
            a+=1
    return new

def print_without_none(matrix):
    for i in matrix:
      for j in i:
        if j!= 0:
          print(j, end='  ')
      print()

n = 5
returned_value = atm_triangle(n)
print_without_none(returned_value)
#This should print
# 1
# 2  2
# 3  4  3
# 4  7  10  4
# 5  11 21  25  5
```

#Trace The BOT

```python
def moving_around(cmds):
  grid = np.full((7, 7), '.')
  x=3
  y=3
  m,n = x,y
  for i in range(len(cmds)):
    if x>-1 and y>-1:
      grid[m][n]="*"
      if cmds[i]==1:
        grid[x][y]='*'
        x-=2
        y-=2
      elif cmds[i]==2:
        grid[x][y]='*'
        x-=2
        y-=1
      elif cmds[i]==3:
        grid[x][y]='*'
        x-=3
        y-=2
      elif cmds[i]==4:
        grid[x][y]='*'
        x-=1
        y-=2
      elif cmds[i]==5:
        grid[x][y]='*'
        x-=2
        y+=1
      elif cmds[i]==6:
        grid[x][y]='*'
        x-=2
        y+=3
      elif cmds[i]==7:
        grid[x][y]='*'
        x-=3
        y+=2
      elif cmds[i]==8:
        grid[x][y]='*'
        x-=1
        y+=2
      elif cmds[i]==9:
        grid[x][y]='*'
        x+=2
        y-=3
      elif cmds[i]==10:
        grid[x][y]='*'
        x+=2
        y-=2
      elif cmds[i]==11:
        grid[x][y]='*'
        x+=1
        y-=2
      elif cmds[i]==12:
        grid[x][y]='*'
        x+=3
        y-=2
      if x>-1 and y>-1:
        m,n = x,y
      grid[m][n]='/'
    else:
      break
    grid[3][3]='-'
  return grid



cmds = np.array([5,11,2,9])
result = moving_around(cmds)
print_matrix(result)
#This should print
# -------------------------------------------
# |  .  |  /  |  .  |  .  |  .  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  .  |  .  |  *  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  *  |  .  |  .  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  .  |  -  |  .  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  .  |  .  |  .  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  .  |  .  |  .  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  .  |  .  |  .  |  .  |  .  |
# -------------------------------------------
```
