```python
class Patient:
  def __init__(self,id=None,n=None,age=None,bld=None):
    self.id=id
    self.name=n
    self.age=age
    self.blood=bld
    self.next=None
    self.prev=None

class WRM:
  def __init__(self):
    self.dh=Patient()
    self.dh.next=self.dh
    self.dh.prev=self.dh
    self.tail=self.dh
  def RegisterPatient(self,id,n,a,b):
    n=Patient(id,n,a,b)

    flag=False
    temp=self.dh.next
    while temp!=self.dh:
      if n.id==temp.id:
        flag=True
      temp=temp.next
    if flag==False:
      n.next=self.dh
      n.prev=self.tail
      self.tail.next=n
      self.tail=n
      self.dh.prev=self.tail
      print('Patient added successfully')
    else:
      print('Could not add patient with same ID.')
  def ServePatient(self):
    if self.dh.next==self.dh==self.dh.prev:
      print('No more patient left to serve')
    else:
      rmv=self.dh.next
      self.dh.next=rmv.next
      rmv.next.prev=self.dh
      elem=rmv.name
      rmv.next=None
      rmv.prev=None
      print(f'{elem} being served.')
  def CancelAll(self):
    if self.dh.next==self.dh==self.dh.prev:
      print('All patients served already.')
    else:
      self.dh=Patient()
      self.dh.next=self.dh
      self.dh.prev=self.dh
      self.tail=self.dh
      print('All appointment cancelled successfully.')
  def CanDoctorGoHome(self):
    if self.dh.next==self.dh==self.dh.prev:
      print('Can Doctor Go Home : YES')
    else:
      print('Can Doctor Go Home : NO')
  def ShowAllPatient(self):
    if self.dh.next==self.dh==self.dh.prev:
      print('No patient left to show.')
    else:
      print('Name of the remaining patients: ',end='')
      temp=self.dh.next
      while temp!=self.dh:
        print(temp.name,end=" ")
        temp=temp.next
      print()

  def ReverseTheLine(self):
    if self.dh.next == self.dh == self.dh.prev:
      print('No patients in the waiting room.')
      return False
    else:
      temp = self.dh.next
      prev_patient = self.dh
      next_patient = temp.next
      while temp != self.dh:
        temp.next = prev_patient
        prev_patient = temp
        temp = next_patient
        if next_patient:
          next_patient = next_patient.next
        else:
          break
      self.tail = self.dh.next
      self.dh.next = prev_patient
      self.tail.prev = self.dh
      self.tail.next = self.dh

      print('Patient line reversed successfully.')
      return True

print('Welcome to waiting room management system.\nChoose your option')
print(f'1.Register Patient\n2.ServePatient\n3.CancelAll\n4.CanDoctorGoHome\n5.ShowAllPatient\n6.ReverseTheLine\n7.Exit')
p1=WRM()
while True:
  choice=int(input('Enter your option:'))
  if choice==1:
    print('Executing RegisterPatient()')
    id=int(input('Enter ID:'))
    n=input('Enter Name:')
    a=int(input('Enter age:'))
    b=input('Enter Blood Group:')
    p1.RegisterPatient(id,n,a,b)
  elif choice==2:
    print('Executing ServePatient()')
    p1.ServePatient()
  elif choice==3:
    print('Executing CancelAll()')
    p1.CancelAll()
  elif choice==4:
    print('Executing CanDoctorGoHome()')
    p1.CanDoctorGoHome()
  elif choice==5:
    print('Executing ShowAllPatient()')
    p1.ShowAllPatient()
  elif choice == 6:
    print('Executing ReverseTheLine()')
    success = p1.ReverseTheLine()
    if success:
      print('Success')
    else:
      print('Not Success')
  elif choice==7:
    print('Thank You for using waiting room management system.')
    break
  else:
    print('No such option')
```
