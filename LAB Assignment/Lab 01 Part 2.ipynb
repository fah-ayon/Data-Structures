```python
import numpy as np
def mean (arr):
  sum= 0
  for i in range(len(arr)):
    sum+= arr[i]
  mean_val = sum/len(arr)
  return mean_val

def sd(arr):
  mean_1 = mean(arr)
  sum_1 = 0
  for i in range(len(arr)):
    sum_1 += (arr[i]-mean_1)**2
  variance = sum_1/len(arr)
  SD= (variance)**0.5
  return(SD)

def away(arr):
  mean_2 = mean(arr)
  sd_1 = sd(arr)
  count= 0
  for i in arr:
    if i> mean_2 + (sd_1 * 1.5) or i < mean_2-(sd_1*1.5):
      count+=1
      n_arr =np.zeros(count, dtype=int)
  if len(n_arr)==0:
    return None
  x= 0
  for j in arr:
    if j> mean_2 + (sd_1 * 1.5) or j < mean_2-(sd_1*1.5):
      n_arr[x]=j
      x+=1
    if x==count:
      return np.array(n_arr)


arr = [10, 20, 30, 40, 50]
arr= np.array(arr)
print('Ques-1 *Dertermine the mean value*')
print(mean(arr))
print('Ques-2 *Standard Deviation*')
print(sd(arr))
print('Ques-3 *Away Values*')
print(away(arr))
```
