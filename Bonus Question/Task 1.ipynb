```python
class vehicleNodes:
  def __init__(self, brand, vehicle_type, rent, passenger, next = None):
    self.brand = brand
    self.vehicle_type = vehicle_type
    self.rent = rent
    self.passenger = passenger
    self.next = next
```

```python

class VehicleHashTable:
  def __init__(self, size):
    self.vehicleTable = [None]*size
    self.empty_slot = {}
    for i in range(size): #{0:False,1:False,2:False,3:false}
      self.empty_slot[i] = False

  def create_from_vehicle_info_array(self, arr):
    for i in arr: #('Toyota', 'Private Car', 500, 4)
      self.insert_vehicle(i)

  def print_vehicle_hashtable(self):
    idx = 0
    for i in self.vehicleTable:
      print(idx, ':', end = ' ')
      head = i
      while head != None:
        print(f'(Brand: {head.brand}, Type: {head.vehicle_type}, Rent: {head.rent}, No. of Passengers: {head.passenger})', end = '---->')
        head = head.next
      print('None')
      print()
      idx += 1

  def find_empty_slot(self):
    for k,v in self.empty_slot.items():
      idx = k
      del self.empty_slot[k]
      return idx




  #Do it yourself
  def __hash_function(self, brand): #Toyota
      index= self.helper(brand,0)%len(self.vehicleTable)
      if index in self.empty_slot:
        del self.empty_slot[index]
      return index
  def helper(self,string,Sum):
    if string=="":
      return Sum
    Sum+=ord(string[0])
    return self.helper(string[1:],Sum)

  #Do it yourself

  def insert_vehicle(self, vehicle):
    hashindex=self.__hash_function(vehicle[0])
    if self.vehicleTable[hashindex]==None:
      node=vehicleNodes(vehicle[0],vehicle[1],vehicle[2],vehicle[3])
      self.vehicleTable[hashindex]=node
      self.empty_slot[hashindex]=node
    else:
      node=vehicleNodes(vehicle[0],vehicle[1],vehicle[2],vehicle[3])
      if node.brand==self.vehicleTable[hashindex].brand:
        node.next= self.vehicleTable[hashindex]
        self.vehicleTable[hashindex]=node
      else:
        for k,v in self.empty_slot.items():
          if v==False:
            self.vehicleTable[k]=node
            self.empty_slot[k]=node
            break
```
